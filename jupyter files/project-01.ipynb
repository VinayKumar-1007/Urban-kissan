```python
from sklearn.metrics import f1_score,plot_confusion_matrix
```

```python
def training(algo,x,y,ratio):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=ratio)
    algo.fit(x_train,y_train)
    pred=algo.predict(x_test)
    plot_confusion_matrix(algo,x_test,y_test)
    score=f1_score(pred,y_test)
    return score
    
    
```

```python
ratios=[0.3,0.2,0.1]
feature=['SFS','SBS','XGB','PCA']
for i in range(0,4):
    for j in range(0,3):
        print("RandomForest with split ratio : {} and feature:{} has accuracy: {}".format(ratios[j],feature[i],training(pipe_rf,X[i],Y[i],ratios[j])))
```

TRAINING WITH NEURAL NETWORK

```python
import tensorflow as tf
```

```python
from keras import models
from keras import layers
```

```python
dat2=dat2.sample(frac=1).reset_index(drop=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X=dat2.loc[:,dat2.columns!='target']
Y=dat2['target']
```

```python
mm=MinMaxScaler()
X=mm.fit_transform(X)
xxgb=mm.fit_transform(xxgb)
```

Plotting Neural network with original data i.e. with all columns and varying epochs

```python
dx=[X,xxgb,x_pca]
dy=[Y,yxgb,y]
nf=[8,12,16]
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
x_val=x_train[:100]
y_val=y_train[:100]
x_train2=x_train[100:]
y_train2=y_train[100:]
for j in nf:
    model=models.Sequential()
    model.add(layers.Dense(32,activation='relu',input_shape=(X.shape[1],)))
    model.add(layers.Dense(j,activation='relu'))
    model.add(layers.Dense(1,activation='sigmoid'))
    model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
    history=model.fit(x_train2,y_train2,epochs=10,batch_size=20,validation_data=(x_val,y_val))
    history_dict=history.history
    acc_values=history_dict['accuracy']
    val_acc_values=history_dict['val_accuracy']
    epochs=range(1,len(acc_values)+1)
    plt.plot(epochs,acc_values,'bo',label='Training_acc')
    plt.plot(epochs,val_acc_values,'b',label='validation_acc')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
    
```

```python
results=model.evaluate(x_test,y_test)
```
