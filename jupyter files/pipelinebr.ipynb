```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("breast cancer")
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop(["Unnamed: 32"],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
np.unique(data["diagnosis"])
```

```python
dummy=pd.get_dummies(data["diagnosis"])
dummy.head()
```

```python
datas=pd.concat([data,dummy],axis=1)
datas.head()
```

```python
datas.drop(["diagnosis"],axis=1,inplace=True)
datas.head()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(data["diagnosis"].values)
Y
```

```python
X=data.drop("diagnosis",axis=1)
X
```

```python
x=X.values
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,Y,stratify=Y)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(StandardScaler(),PCA(n_components=2),LogisticRegression())
pipe.fit(x_train,y_train)
```

```python
u=pipe.predict(x_test)
u
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(u,y_test)*100
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(estimator=pipe,X=x_train,y=y_train,cv=10,n_jobs=-1)
```

```python
print('cv accuracy scores=%s',scores)
```

```python
print(np.mean(scores),np.std(scores))
```

```python
from sklearn.model_selection import learning_curve
from sklearn.linear_model import LogisticRegression
```

```python
pipe_lr=make_pipeline(StandardScaler(),LogisticRegression(penalty='l2',random_state=1))
pipe_lr.fit(x_train,y_train)
```

```python
train_sizes,training_scores,testing_scores=learning_curve(estimator=pipe_lr,X=x_train,y=y_train,train_sizes=np.linspace(0.01,1.0,10),cv=10,n_jobs=1)
```
