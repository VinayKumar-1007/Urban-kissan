```python
df2=df2.dropna().reset_index(drop=True)
```

```python
df2=df2.sample(frac=1).reset_index(drop=True)
x=df2.iloc[:,df2.columns!='n_cores']
y=df2['n_cores']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
ssl=StandardScaler()
```

```python
x=ssl.fit_transform(x)
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x,y)
```

```python
i=10
x2=[]
while(i<2000):
    x2.append(data.iloc[i,data.columns!='n_cores'].values)
    i+=10
```

```python
df=pd.DataFrame(data=x2)
```

```python
xs=df.iloc[:,df.columns]
```

```python
xs=ssl.fit_transform(xs)
```

```python
s=rf.predict(xs)
```

```python
from sklearn.metrics import accuracy_score
```

```python
res=pd.DataFrame(data=s,columns=['predicted'])
```

```python
ori=pd.DataFrame(data=original,columns=['original'])
```

```python
final=pd.concat([res,ori],axis=1)
final.head(20)
```

```python
xa=final['predicted'].values
ya=final['original'].values
accuracy_score(xa,ya)
```
